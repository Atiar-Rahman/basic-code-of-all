# matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### bar plot

```python
x = ['python','c','c++','js']
y = [85,78,60,65]
y1 = [8,7,6,6]
c= ['y','b','m','g']
plt.bar(x,y,width=0.5,color=c,align='center',edgecolor='r',linewidth=4,linestyle='--',alpha=0.4,label='c') # default align center
plt.bar(x,y1)
plt.legend()
plt.xlabel('language',fontsize=20)
plt.ylabel('No',fontsize=20)
plt.title("programming",fontsize=20)
plt.show()
```
