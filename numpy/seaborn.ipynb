```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
tips
```

```python
penguins = sns.load_dataset("penguins")
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

```python
sns.pairplot(data=penguins, hue="species")
```

# Data visualization

##### Line plot



```python
var = [1,2,3,4,5,6,7,8]
var_1 = [4,3,5,7,8,9,1,2]


df = pd.DataFrame({'var':var,'var_1':var_1})

sns.lineplot(x=var,y=var_1, data=df)
plt.show()
```

```python
sns.lineplot(x=var,y=var_1)
```

```python
y_1 = sns.load_dataset('penguins')
y_1.head(5)
```

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=y_1, hue='sex',style='sex',palette="",markers=['o','>'],dashes=True)
plt.show()
```

```python
import seaborn as sns

print(sns.color_palette().as_hex())   # default palette (hex codes)
print(sns.palettes.SEABORN_PALETTES.keys())   # list of seaborn palettes
```

```python
# list all available colormaps
print(plt.colormaps())

# filter only those starting with "A"
a_palettes = [cmap for cmap in plt.colormaps() if cmap.lower().startswith("a")]
print(a_palettes)
```
