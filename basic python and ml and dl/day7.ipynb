## What is machine learning?
machine learning is a branch of artificial intelligence that enables computers to learn from data and improve their performance on tasks without being explicity programmed. Instead of following strict rules, ml systems find patterns in data and make decisions or predictions based on those patterns

### types of machine learning
machine learning is boardly categorized into three main types:

- supervised learning
In supervised learning, the model is trained on labeled data, that is, data that includes both the input and the correct output.
Goal: learn a mapping from inputs to outputs so it can predict the output for new unseen inputs.

Example Tasks:
- classificition: predict categories(spam vs not spam)
- Regressation: Predict continious values(house prices)

Exam:
- Email spam detection
- predictig stock price
- diagnosing diseases based on symtoms

## unsupervised learning
In upsupervised learning the model is given unlabeled data and must find structure or patterns within it.
- goal: discover hidden patterns or grouping in data
- Example tasks:
    - clustering: group similar items(customer segmentation)
    - dimensionality reduction: simplify data while preserving important information.(PCA)

example:
- market basket analysis(what products are bought together)
- grouping customers by behavior
- detecting anomalies(fraud detection)


## Reinforcement learning
in reinforcement learning an agent learns to make decisions by interacting with an environment. it receives rewards or penalties for its actions and learns to maximize cumulative rewards over time

- goal: learn a polivy for making decisions that maximize long term reward
- Example tasks
    - game playing(e.g. chess,go,video games)
    - robotics
    - self driving cars

examples:
- AlphaGo by deepmind
- autonomous drone navigation
- personalized recommandations that adapt over time

## simple math
- mean: Aerage of data points.
- median: Middle value of stored data
- variance: Average squared distance from mean
- std deviation: spread of data (sqrt of variance)
- Linear Equation: y = mx+b; used in regression
- probability: chance of an event
- bayes theorem: updates probability based on new evidence
- Gradiet: Slope/derivative, used in optimization.

```python
import statistics  as st
mean = st.mean(arr)
median = st.median(arr)
variance = st.variance(arr)
std_dev = st.stdev(arr)

print(f"mean {mean}")
print(f"median: {median}")
print(f"variance {variance}")
print(f"standard deviation: ",std_dev)
```

```python
data = np.array([3,7,8,10])
mean = np.mean(data)
median = np.median(data)
variance = np.var(data)
std_dev = np.std(data)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
```

```python
import sympy as sp

x = sp.Symbol('x')
f = x**2

# Compute derivative
grad = sp.diff(f, x)
print("Gradient (df/dx):", grad)
```

```python
# Gradient Descent on f(x) = x^2
def f(x):
    return x**2

def grad_f(x):
    return 2 * x

x = 10      # start point
lr = 0.1    # learning rate
epochs = 20

for i in range(epochs):
    grad = grad_f(x)
    x = x - lr * grad
    print(f"Step {i+1}: x = {x}, f(x) = {f(x)}")
```
