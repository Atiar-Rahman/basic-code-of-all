```python
import cv2

# ১. ফেস ডিটেকশন মডেলটি লোড করা
# (এই ফাইলটি সাধারণত OpenCV এর সাথে থাকে, না থাকলে ডাউনলোড করে নিতে হয়)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# ২. ওয়েবক্যাম চালু করা (০ মানে ডিফল্ট ক্যামেরা)
cap = cv2.VideoCapture(0)

while True:
    # ফ্রেম রিড করা
    ret, frame = cap.read()
    
    # কোড সহজ করার জন্য ফ্রেমটিকে গ্রে-স্কেলে রূপান্তর করা
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # ৩. মুখ শনাক্ত করা
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)
    
    # ৪. শনাক্ত করা মুখের চারপাশে চারকোনা বক্স আঁকা
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2) # নীল রঙের বক্স
        
    # ফলাফল দেখানো
    cv2.imshow('Face Detection', frame)
    
    # 'q' চাপলে প্রোগ্রাম বন্ধ হবে
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# সব কিছু বন্ধ করা
cap.release()
cv2.destroyAllWindows()
```
