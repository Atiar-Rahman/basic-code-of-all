# python GrayScaling of images using opencv
GrayScaling is the process of converting an image from other color spaces eg. RGB,CMYK,HSV etc. to shades of gray. it varies between complete black and complete white.

Importance of grayscaling
1. Dimension reduction: for example, in RGB images there are three color channels and three dimensions while grayscale images are single-dimension
2. reduce model complexity: consider training neural articles on RGB image 10X10X3 pixels. the input layer will have 300 imput nodes. on the other hand, the same neural network will need only 100 input nodes for grayscale images.
3. other algorithms to work: many algorithms are customized to work only on grayscale images. canny edge detection function preimplemented in the opencv library works on grayscale images only.

let's learn the different image processing methods to convert a colored image into a grayscale image.

Method 1: using the cv2.cvtColor() function
    - import the opencv and read the original image using imread() than convert to grayscale using cv2.cvtColor() function. destroyAllWindrows() function allows users to destroy or close all windows at any time after existing the script.

Ah! You’re asking about the `0.33` in your manual grayscale conversion. Let’s break it down carefully.

---

### 1️⃣ Why `0.33`?

* In your code:

```python
img[i, j] = sum(img[i, j]) * 0.33
```

* `sum(img[i, j])` adds the **B, G, R values**.
* To get the **average**, you divide by 3:

$$
\text{Average} = \frac{B + G + R}{3} \approx (B + G + R) \times 0.3333
$$

* So `0.33` is **just an approximation of 1/3**.

---

### 2️⃣ Better Approximation

* If you want **more precise averaging**, use:

```python
img[i, j] = sum(img[i, j]) / 3
```

* This avoids rounding errors from `0.33`.

---

### 3️⃣ Human-Perceived Grayscale (Optional)

* Human eyes see **green brighter than red and blue**, so a weighted formula gives better results:

$$
\text{Gray} = 0.114*B + 0.587*G + 0.299*R
$$

* In OpenCV, this is used internally by:

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

---

✅ **Summary:**

* `0.33` ≈ `1/3` → simple average of B, G, R channels.
* Better: divide by 3 or use weighted formula for perceptual accuracy.

---


````python
img = cv2.imread('birds1.jpeg')

(row,col) = img.shape[0:2]

for i in range(row):
    for j in range(col):
        img[i,j] = sum(img[i,j])/3
        
        
plt.imshow(img)
plt.title("Gray scale")
plt.show()
````
